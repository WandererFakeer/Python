## Credit - checks the validity of a given credit card number

```python
def num_list(card):
    
    '''
    
    :param card: 
    :return: nums
    
    This function takes one input as a string, and returns a list of integers
    
    '''
    
    index_1 = 1

    #empty list
    nums = []

    while (index_1 < len(card)):
    
        #appends the list, with integer of every second character of the string, 
         # multiplied by 2
        nums.append(int(card[index_1]) * 2)
    
        index_1 += 2
        
    return nums
```

```python
def adding(nums):
    
    '''
    
    :param nums: 
    :return: add
    
    This function takes a list of integers as an input, and returns an integer
    
    '''
    
    add = 0

    for num in nums:
    
        #checks if each number from the list is greater than 9
        if (num > 9):
            
            #sums the quotient with add
            add += num // 10
        
            #then sums the modulus with add
            add += num % 10
        
        #else only sums the modulus with add
        else:
        
            add += num % 10 
            
    return add 
```

```python
def add_num(card, sums):
    '''

    :param card:
    :param sums:
    :return: sums

    This function takes 2 inputs, a string and an integer, and returns the same integer

    '''
    
    index_2 = 0

    while (index_2 < len(card)):
    
        # sums with integer of every alternate character of the string,
         # with which no calculation was made before
        sums += int(card[index_2]) 
    
        index_2 += 2
        
    return sums
```
