```python
rock_ascii = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper_ascii = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors_ascii = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''
```

```python
import random
```

```python
def graphic (user, who):
    """
            This function takes user choice and who (computer or user) as 2 parameters,
            and prints the choices, according to the user or computer,
            with values from the ascii arts

    """

    if (user == "rock"):

        print(f"{who} chose: {rock_ascii}")

    elif (user == "paper"):

        print(f"{who} chose: {paper_ascii}")

    else:
        print(f"{who} chose: {scissors_ascii}")
```

```python
def result (user_result, computer_result):
    """
        This function takes user choice and computer random choice, as 2 parameters, and
        checks the wining conditions, with "If-elif"

    """

    if (user_result == computer_result):

        print("Draw")

    elif ((user_result == "rock" and computer_result == "scissors") or
        (user_result == "paper" and computer_result == "rock") or
        (user_result == "scissors" and computer_result == "paper")):

        print("You win!")

    elif ((user_result == "rock" and computer_result == "paper") or
        (user_result == "paper" and computer_result == "scissors") or
        (user_result == "scissors" and computer_result == "rock")):

        print("You lose!")
        
    else:
        print(f"You entered {user_result}. It's an invalid entry")
```

```python
def the_game ():

    game = ["Rock", "Paper", "Scissors"]

    user_choice = input('What do you choose? Type "Rock", "Paper" or "Scissors":\n').lower()

    graphic(user_choice, "You")

    computer_choice = random.choice(game).lower()

    graphic(computer_choice, "Computer")

    if user_choice.title() in game:

        result(user_choice, computer_choice)

    else:
        print("You chose wrong option.")
```

```python
while True:

    the_game()

    continuity = input('Do you want to play again? Type "Yes" or "No". ').lower()

    if (continuity == "no"):

        break
```
