## Cash - Amount of Coins needed to pay

```python
def amount(money, denomination):

    '''

    :param money:
    :param denomination:
    :return: tuple of remaining amount and the count

    This function takes 2 integers, the money amount and the denomination and
    returns the tuple of remaining amount and the count
    '''

    #count variable keeps track of the number of coins
    count = 0

    #Checks if money amount is greater than or equal to the denomination
    if (money >= denomination):

        count = money // denomination

        #The remaining amount is calculated by subtracting the amount,
         #which denomination cleanly divides
        money -= (count * denomination)

        #returns the tuple of remaining amount and the count
        return money, count

    elif (money == 0):

        return 0, 0

    else:

        return money, count
```

```python
def coin_need():

    while True:

        total = int( input ("Enter the Amount - "))

        if (total > 0):

            break

    outstanding_1, sum_1 = amount(total, 25)

    outstanding_2, sum_2 = amount(outstanding_1, 10)

    outstanding_3, sum_3 = amount(outstanding_2, 5)

    outstanding_4, sum_4 = amount(outstanding_3, 1)

    total_count = (sum_1 + sum_2 + sum_3 + sum_4)

    print(f"It will take {total_count} coins to return {total}")
```

```python
while True:
    
    cont = input ("Do you want to check how many coins you need to pay? 'Yes' or 'No'? ")
    
    if (cont.lower() == "no"):

        break

    coin_need()
```
