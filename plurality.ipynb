## Plurality - Implement a program that runs a plurality election

```python
def candidate_list():
    """

    :return: candi_list

    This function returns a list of string

    """
    candi_num = int(input("How many candidates are there? "))

    if (candi_num <= MAXIMUM_CANDIDATES):
        print("Enter the names of candidates:")

        # The list is populated with user input about the number of candidates
        candi_list = [input() for _ in range(candi_num)]

        return candi_list
```

```python
def valid_votes(voters_num):

    """

    :param voters_num:
    :return: peoples_votes

    This function takes an integer and returns a list of valid votes, which is a list of strings

    """

    print("Enter the votes of voters:")

    peoples_votes = []

    for _ in range(voters_num):

        #asks the user for votes
        response = input("Vote: ")

        #checks if voted name is in the list of candidates
        if (response in candidates):

            peoples_votes.append(response)

        else:

            continue

    return peoples_votes
```

```python
def winner(voted_list):

    """

    :param voted_list:
    :return: name

    This function takes one input as a list of string, and returns a string from the list with maximum number
    """

    #creates a dictionary with voted names as keys and voted numbers as value
    candidates_votes = {voted_list[index]: voted_list.count(voted_list[index]) for index in range(len(voted_list))}

    #gets the maximum number from the dictionary with max() function
    max_vote = max(candidates_votes.values())

    #empty string
    name = ""

    for vote in candidates_votes:

        #checks if the number of votes equals with the maximum value
        if (candidates_votes[vote] == max_vote):

            #appends the name of the candidate with the maximum number of votes to variable name
            name += vote + " "

    return name
```

```python
#Maximum number of allowed candidates - Constant
MAXIMUM_CANDIDATES = 9

#calls the "candidate_list()" function and saves the returned list of candidates
candidates = candidate_list()

#asks the user about the number of voters
number_of_voters = int(input("How many voters are there? "))

#calls the function "valid_votes()", passing variable "number_of_voters" and returns a list of votes
peoples_votes = valid_votes(number_of_voters)

#calls the function "winner()", passing the list "peoples_votes", and returns a string
winner_name = winner(peoples_votes)

print(f"The winner/s is/are - {winner_name}")
```
