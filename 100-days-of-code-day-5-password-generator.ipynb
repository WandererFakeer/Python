```python
import random
```

```python
def alphabets():

    #empty list to be filled with the capital and small alphabets
    alphabets_list = []

    #list comprehension with the ASCII values of capital alphabets
    capital = [chr(num) for num in range(65, 91)]

    # list comprehension with the ASCII values of small alphabets
    small = [chr(num) for num in range(97, 123)]

    #extending the empty list with capital and small alphabets lists
    alphabets_list.extend(capital)

    alphabets_list.extend(small)

    return alphabets_list
```

```python
def symbols():

    # list comprehension with the ASCII values of symbols
    symbol_list = [chr(num) for num in range(35, 39)]

    #extra symbols list, namely "!", "@" and "*"
    extra = [chr(33), chr(42), chr(64)]

    #extend the symbols list with the extra symbols list
    symbol_list.extend(extra)

    return symbol_list
```

```python
def numbers():

    # list comprehension with the ASCII values of integers from 0 to 9
    num_list = [chr(num) for num in range(48, 58)]

    return num_list
```

```python
def pass_choice(appropriate_list, chosen_num):

    """

    :param appropriate_list:
    :param chosen_num:
    :return: password_list

    This function takes 2 parameters, as an integer number and a list, and
    appends the values from the passed list, till it reaches the range of the
    passed integer value.
    It returns the appended new list

    """

    random_generation_list = []

    #loop through the given range of number
    for index in range(chosen_num):

        #randomly choose a value from the passed list, and append it to the empty list
        random_generation_list.append(random.choice(appropriate_list))

    return random_generation_list
```

```python
def password_list():

    random_chosen_password_list = []

    letters_num = int(input("How many letters would you like in your password?\n"))

    #call the "alphabets()" function to return the list of all the alphabets
    letters_list = alphabets()

    #call "pass_choice()", passing the alphabet list and user chosen number of alphabets to it
    chosen_alphabets_list = pass_choice(letters_list, letters_num)

    #add the randomly chosen letters to the list "random_chosen_password_list"
    random_chosen_password_list += chosen_alphabets_list

    symbols_num = int(input("How many symbols would you like in your password?\n"))

    #call the "symbols()" function to return the list of all the symbols
    symbols_list = symbols()

    # call "pass_choice()", passing the symbols list and user chosen number of symbols to it
    chosen_symbols_list = pass_choice(symbols_list, symbols_num)

    # add the randomly chosen symbols to the list "random_chosen_password_list"
    random_chosen_password_list += chosen_symbols_list

    integers_num = int(input("How many numbers would you like in your password?\n"))

    #call the "numbers()" function to return the list of the from 0 to 9
    integers_list = numbers()

    # call "pass_choice()", passing the numbers list and user chosen number of numbers to it
    chosen_numbers_list = pass_choice(integers_list, integers_num)

    # add the randomly chosen numbers to the list "random_chosen_password_list"
    random_chosen_password_list += chosen_numbers_list

    #shuffle the whole randomly chosen list to jumble the elements
    random.shuffle(random_chosen_password_list)

    #convert the list to a string
    password = "".join(random_chosen_password_list)

    return password
```

```python
print("Welcome to the PyPassword Generator!")

user_password = password_list()

print(f"Your password is: {user_password}")
```
