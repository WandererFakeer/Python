# Create a Tip Calculator, based on the Total Bill, Percentage of Tip, and the Number of Individuals

```python
def tip_calculator (bill, percentage, number):
    
    '''
        This function takes 3 integer parameters, and calculates the individual tip amount
    
    '''
    
    #1. Diving the percentage of tip by 100, we get decimal tip amount, and add it to 1. Then multiply by the bill amount to get the total amount of bill. Then dividing the total amountof bill, to find the individual amount of bill
    
    individual_bill = ( bill * ( 1 + (percentage / 100) ) ) / number 
    
    #2. Format the bill amount to 2 precision points
    individual_bill = "{:.2f}".format(individual_bill)
    
    #3. Returns the individual bill
    
    return individual_bill
```

```python
print("Welcome to the Tip Calculator")

#4. Ask the user how much is the bill
bill_amount = float ( input ("What is the amount of bill? $" ) )

#5. Ask the user about the percentage of tip among 10%, 12% ad 15%
tip_percentage = int ( input ("What percentage of tip would you like to give? ") )

#6. Ask the user how many individual are there
individual = int ( input ("How many people to share the bill? ") )

#7. Call the function "tip_calculator()", passing bill amount, tip percentage and number of people as 3 arguments. Save the returned value to "amount"
amount = tip_calculator (bill = bill_amount, percentage = tip_percentage, number = individual)

print(f"Every person should pay : ${amount}")
```
