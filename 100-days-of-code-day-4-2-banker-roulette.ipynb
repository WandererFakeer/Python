**Write a program that will select a random name from a list of names. The person selected will have to pay for everybody's food bill.**

```python
#1. Import the random module
import random

#2. Define a function "random_name()", which takes one parameter as a list, and returns the randomly selected item from the given list
def random_name (names):
    
    person = random.choice(names)
    
    return person
```

```python
#3. Ask the user to type their friends' names, followed with a comma and space. Then split those user inputs with ", " delimiter, which converts the String to a List

names_list = input ("Enter your friends' names and separate each name with a comma and space, i.e. (, )\n").split(", ")

random_person = random_name(names = names_list)

print(f"{random_person} is going to buy the meal today!")
```
