```python
from IPython.display import clear_output
```

```python
logo = '''
                         ___________
                         \         /
                          )_______(
                          |"""""""|_.-._,.---------.,_.-._
                          |       | | |               | | ''-.
                          |       |_| |_             _| |_..-'
                          |_______| '-' `'---------'` '-'
                          )"""""""(
                         /_________\\
                       .-------------.
                      /_______________\\
'''
```

```python
def auction(user_name, user_bid):

    """

    :param user_name:
    :param user_bid:
    :return: auction_dictionary

    This function takes 2 parameters, one string "user_name" and one float "user_bid",
    and returns dictionary "auction_dictionary", populated with "user_name" as key and
    "user_bid" as value

    """
    auction_dictionary[user_name] = user_bid

    return auction_dictionary
```

```python
def winner():

    #gets the highest amount of bid with "max()" function, passing the values of the dictionary
    max_bid_amount = max(auction_dictionary.values())

    #extracts the name of the highest bidder, from the value of the dictionary
    max_bidder = list(auction_dictionary.keys())[list(auction_dictionary.values()).index(max_bid_amount)]

    return max_bidder, max_bid_amount
```

```python
def auction_program():

    while True:

        name = input("What is your name? ")

        amount = float(input("What is your bid? $"))

        #calls the function "auction(), passing "name" and "amount" as arguments
        auction(name, amount)

        more_bidder = input("Are there any other bidders? Type 'Yes' or 'No': ").lower()
        
        if (more_bidder == "yes"):
            
            clear_output()

        elif (more_bidder == "no"):

            #calls "winner()", which returns the bid winner, with their bid value
            bid_winner, bid_amount = winner()

            print(f"The winner is: {bid_winner} with an amount of ${bid_amount}")

            break
```

```python
print(logo)

auction_dictionary = {}

auction_program()
```
