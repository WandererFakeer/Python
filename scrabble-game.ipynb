## Scrabble

```python
from IPython.display import clear_output
```

```python
def structure():
    """

    :return: a dictionary of letters and points

    """
    #creates a list with all the English alphabets in capital letters
    alphabets_list = [chr(index) for index in range(65, 91)]

    points = [1, 3, 3, 2, 1, 4, 2, 4, 1, 8, 5, 1, 3, 1, 1, 3, 10, 1, 1, 1, 1, 4, 4, 8, 4, 10]

    #creates a dictionary of alphabets and points, with dictionary comprehension
    alphabets_dictionary = {alphabets_list[index]: points[index] for index in range(len(alphabets_list))}

    return alphabets_dictionary
```

```python
def total_score(alphabets, word):

    """
    :param alphabets:
    :param word:
    :return: sum of points

    """
    sum = 0

    for char in word:

        sum += alphabets[char]

    return sum
```

```python
def game():

    #calls "structure()", which returns a dictionary of English letters and points
    alphabets = structure()

    #asks the user to enter a word
    player_1 = input("Player 1 : Enter a Word : ").upper()
    
    clear_output()

    player_2 = input("Player 2 : Enter a Word : ").upper()

    #calls "total_score()", with the returned dictionary "alphabets" and "player_1" as inputs
    score_1 = total_score(alphabets, player_1)

    score_2 = total_score(alphabets, player_2)

    if (score_1 > score_2):

        print(f"Player 1 scored {score_1}, Player 2 scored {score_2}. So, Player 1 Won the game")

    elif (score_1 < score_2):

        print(f"Player 1 scored {score_1}, Player 2 scored {score_2}. So, Player 2 Won the game")
        
    else:
        
        print(f"Player 1 and Player 2 both scored {score_1}. So it is Tie!")
```

```python
while True:

    game()

    response = input("Do you want to play again? 'Yes' or 'No'? ")

    if response.lower() == "no":

        break
```
