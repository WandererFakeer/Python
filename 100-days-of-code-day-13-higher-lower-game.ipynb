```python
from IPython.display import clear_output

import random
```

```python
def choice_random():

    """

    :return: choice

    This function returns a random dictionary from "data" list

    """
    choice = random.choice(data)

    return choice
```

```python
def format_code(account):

    """
    Format account into printable format: name, description and country

    """

    name = account["name"]

    description = account["description"]

    country = account["country"]

    return f"{name}, a {description}, from {country}"
```

```python
def game():

    print(logo)

    score = 0

    #first, to choose 2 random dictionary from "data" list, call "choice_random()" twice
    first_random_chosen = choice_random()

    second_random_chosen = choice_random()
    
    #checks if both randomly generated data are same, then regenerates second option
    if first_random_chosen == second_random_chosen:
        
        second_random_chosen = choice_random()

    while True:

        print(f"Compare A: {format_code(first_random_chosen)}")

        print(vs)

        print(f"Against B: {format_code(second_random_chosen)}")

        #saves the value of "follower_count" for the 2 chosen dictionary
        first_follower = first_random_chosen['follower_count']

        second_follower = second_random_chosen['follower_count']

        user_choice = input("Who has more followers? Type 'A' or 'B': ").lower()
        
        clear_output()

        #checks if user chose "a", which corresponds to the "follower_count"
         #of the first dictionary, and also checks if first "follower_count" is
          #greater or equal to the second "follower_count".
           #or if user chose "b", which corresponds to the "follower_count"
            #of the second dictionary, and also checks if second "follower_count" is
             #greater or equal to the first "follower_count".
        if ((user_choice == "a" and first_follower >= second_follower) or
                (user_choice == "b" and second_follower >= first_follower)):

            #adds 1 to "score"
            score += 1

            #assigns the second chosen dictionary to "first_random_chosen"
            first_random_chosen = second_random_chosen

            #calls "choice_random()" to randomly choose another dictionary and
             #assigns the dictionary to "second_random_chosen"
            second_random_chosen = choice_random()

            print(f"You're right! Current score: {score}")

        #else if user chose "a", which corresponds to the "follower_count"
         #of the first dictionary, and also checks if first "follower_count" is less than
          #the second "follower_count".
           #or if user chose "b", which corresponds to the "follower_count"
            #of the second dictionary, and also checks if second "follower_count" is less than
             #the first "follower_count".
        elif ((user_choice == "a" and first_follower < second_follower) or
              (user_choice == "b" and second_follower < first_follower)):

            print(f"Sorry, that's wrong. Final score: {score}")

            break
```
