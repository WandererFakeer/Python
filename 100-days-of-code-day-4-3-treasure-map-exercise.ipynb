**Write a program that will mark a spot with an X.
A variable contains a nested list. When this variable is printed this is what the nested list looks like:
[['⬜️', '⬜️', '⬜️'],['⬜️', '⬜️', '⬜️'],['⬜️', '⬜️', '⬜️']]
Take the user input and use that input to update nested list with an "x"
The first digit in the input will specify the column (the position on the horizontal axis).
The second digit in the input will specify the row number (the position on the vertical axis)**

```python
#1. Create 3 separate lists
map_1 = ['⬜️', '⬜️', '⬜️']

map_2 = ['⬜️', '⬜️', '⬜️']

map_3 = ['⬜️', '⬜️', '⬜️']

#2. Take one list and nest other 3 lists inside this list, to make it a nested list
map_full = [
    map_1, 
    map_2, 
    map_3
]

print (f"Before putting your marker, the map looks like - \n{map_1}\n{map_2}\n{map_3}")
```

```python
#3. Ask the user to enter a position to put a marker
user_mark = input ("Enter a place, where you want to put your marker : ") 

#4. Consider the first digit to be the position of the column, and convert it to int()
column = ( int (user_mark[0]) - 1) 

#5. Consider the second digit to be the position of the row, and convert it to int()
row = (int (user_mark[1]) - 1)

#6. Access the position of the Row and Column on the nested list, and put a marker "X" on that position
map_full[row][column] = "X"

print (f"After marking your position, the map looks like - \n{map_1}\n{map_2}\n{map_3}")
```
