```python
def addition(first_num, second_num):

    return (first_num + second_num)
```

```python
def subtraction(first_num, second_num):

    return (first_num - second_num)
```

```python
def multiplication(first_num, second_num):

    return (first_num * second_num)
```

```python
def division(first_num, second_num):

    return (first_num / second_num)
```

```python
def calculator():

    print(logo)

    first = float(input("What's the first number?: "))

    while True:

        operations = {
            "+": addition,

            "-": subtraction,

            "*": multiplication,

            "/": division
        }

        for operation in operations:

            print(operation)


        operator = input("Pick an operation: ")

        second = float(input("What's the next number?: "))

        #object reference to operations
        user_operation = operations[operator]

        #calls the appropriate function
        result = user_operation(first, second)

        print(f"{first} {operator} {second} = {result}")

        continuity = input(f"Type 'Yes' to continue calculating with {result}, "
                           f"or type 'No' to start a new calculation: ").lower()

        if continuity == "yes":

            first = result

        if continuity == "no":

            break
```

```python
while True:

    calculator()
```
