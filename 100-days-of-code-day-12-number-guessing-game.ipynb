# Number Guessing Game Objectives:

**Include an ASCII art logo.
Allow the player to submit a guess for a number between 1 and 100.
Check user's guess against actual answer. Print "Too high." or "Too low." depending on the user's answer.
If they got the answer correct, show the actual answer to the player.
Track the number of turns remaining.
If they run out of turns, provide feedback to the player.
Include two different difficulty levels (e.g., 10 guesses in easy mode, only 5 guesses in hard mode).**

```python
logo = """                                                                                                                                        

  _  _            _                ___                _              ___                
 | \| |_  _ _ __ | |__  ___ _ _   / __|_  _ ___ _____(_)_ _  __ _   / __|__ _ _ __  ___ 
 | .` | || | '  \| '_ \/ -_) '_| | (_ | || / -_|_-<_-< | ' \/ _` | | (_ / _` | '  \/ -_)
 |_|\_|\_,_|_|_|_|_.__/\___|_|    \___|\_,_\___/__/__/_|_||_\__, |  \___\__,_|_|_|_\___|
                                                            |___/                       
"""
```

```python
import random
```

```python
#randomly choose a number, between 1 and 100
number = random.randint(1, 100)
```

```python
def guess_game(living):

    print(f"You have {living} attempts remaining to guess the number.")

    while living != 0:

        guess = int(input("Make a guess: "))

        #checks if guess is not equal to number
        if guess != number:

            #decrease "living" by 1
            living -= 1

            #checks if "living" is equal to 0
            if living == 0:

                print(f"You've run out of guesses, you lose.")

            #if "living" not equal to 0, checks if guess is greater than the number
            elif guess > number:

                print(f"Too high.\nGuess again.\nYou have {living} attempts remaining "
                  "to guess the number.")

            else:

                print(f"Too low.\nGuess again.\nYou have {living} attempts remaining "
                  "to guess the number.")

        #checks if guess is equal to the number
        else:

            print(f"You got it! The answer was {number}.")

            break
```

```python
print(logo)

print("Welcome to the Number Guessing Game!\nI'm thinking of a number between 1 and 100.")

difficulty = input("Choose a difficulty. Type 'Easy' or 'Hard': ").lower()

if difficulty == "easy":

    lives = 10

    guess_game(lives)

else:

    lives = 5

    guess_game(lives)
```
