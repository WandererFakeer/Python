```python
def fill_blanks(chosen, user, living):

    """

    :param chosen:
    :param user:
    :param living:
    :return: word_space, living

    This function takes 3 parameters, as 2 strings, one as the randomly chosen word
    and another as letter chosen by the user, and 1 integer, which keeps track of the lives,
    and returns the word filled list and remaining lives

    """

    #checks if the letter (user) chosen by the user is not in the random word (chosen)
    if user not in chosen:

        #living decreases by 1
        living -= 1

        print(f"You chose: {user}. It's a wrong letter")

    else:

        random_chosen_length = len(chosen)

        #traverses through the length of the random word
        for index in range(random_chosen_length):

            #if user letter equals the random word, index wise
            if user == chosen[index]:

                #the blank list gets filled with the letter at the exact index
                word_space[index] = user

    return word_space, living
```
