## Blackjack House Rules

**The deck is unlimited in size.**
**There are no jokers.**
**The Jack/Queen/King all count as 10.**
**The Ace can count as 11 or 1.**
**Use the following list as the deck of cards:**
**cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]**
**The cards in the list have equal probability of being drawn.**
**Cards are not removed from the deck as they are drawn.**
**The computer is the dealer.**

```python
from IPython.display import clear_output

import random
```

```python
def deal_card():

    """
    :return: random_card

    This function that uses a List, "cards" to return a random card from that list
    11 is the Ace.

    """

    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

    random_card = random.choice(cards)

    return random_card
```

```python
def calculate_score(cards_list):

    """

    :param cards_list:
    :return: sum(cards_list)

    This function takes a list, "cards_list" as a parameter, and
    returns the sum of the elements of the list

    """

    #check for a blackjack (a hand with only 2 cards: ace + 10) and return 0
     #instead of the actual score.
    if 11 in cards_list and sum(cards_list) == 21:

        return 0

    #Check for 11 (ace). If the score is already over 21, remove 11 and replace it with 1.
    elif 11 in cards_list and sum(cards_list) > 21:

        cards_list.remove(11)

        cards_list.append(1)

    return sum(cards_list)
```

```python
def compare(u_score, com_score):
    #Create a function called compare() and pass in the user_score and computer_score

    #If the computer and user both have the same score, then it's a draw
    if u_score == com_score:

        print("Draw ðŸ™ƒ")

    #If the user has a blackjack (0), then the user wins.
    elif u_score == 0:

        print("You hit Blackjack! You win ðŸ˜Ž")

    #If the computer has a blackjack (0), then the user loses
    elif com_score == 0:

        print("Opponent has Blackjack! You lost ðŸ˜±")

    #If the user_score is over 21, then the user loses.
    elif u_score > 21:

        print("You went over. You lose ðŸ˜­")

    #If the computer_score is over 21, then the computer loses.
    elif com_score > 21:

        print("Opponent went over. You win ðŸ˜")

    #If none of the above, then the player with the highest score wins.
    elif u_score > com_score:

        print("You win ðŸ˜ƒ")

    else:

        print("You lose ðŸ˜¤")
```

```python
def game():

    print(logo)

    #Deal the user 2 cards using deal_card() from Day_11_Capstone_Project_Blackjack_Functions
    user_cards = [deal_card() for _ in range(2)]

    #to calculate user's score, call "calculate_score()" from
     #Day_11_Capstone_Project_Blackjack_Functions, passing "user_cards" as argument
    user_score = calculate_score(user_cards)

    print(f"Your cards: {user_cards}, Your score: {user_score}")

    # Deal the computer 2 cards using deal_card() from Day_11_Capstone_Project_Blackjack_Functions
    computer_cards = [deal_card() for _ in range(2)]

    # to calculate computer's score, call "calculate_score()" from
    # Day_11_Capstone_Project_Blackjack_Functions, passing "computer_cards" as argument
    computer_score = calculate_score(computer_cards)

    print(f"Computer's cards: {computer_cards[0]}")

    #Call calculate_score(). If the computer or the user has a blackjack (0) or if
     #the user's score is over 21, then the game ends.
    while (user_score != 0) and (user_score < 21) and (computer_score != 0) and (computer_score != 21):

        #If the game has not ended, ask the user if they want to draw another card.
        draw_another = input("Type 'Y' to get another card, type 'N' to pass: ").lower()

        #If yes, then use the deal_card() function to add another card to the user_cards List.
        #If no, then the game has ended.
        if draw_another == "y":

            user_cards.append(deal_card())

            user_score = calculate_score(user_cards)

            if user_score != 21:

                print(f"Your cards: {user_cards}, Your score: {user_score}")

        else:

            #once the user is done, it's time to let the computer play.
             #the computer should keep drawing cards as long as it has a score less than 17.
            while computer_score < 17:

                computer_cards.append(deal_card())

                computer_score = calculate_score(computer_cards)

            else:

                break

    print(f"Your final cards: {user_cards}, Your final score: {user_score}")

    print(f"Computer's final cards: {computer_cards}, Computer's final score: {computer_score}")

    compare(user_score, computer_score)
```

```python
while True:

    play = input("Do you want to play a game of Blackjack? Type 'Yes' or 'No': ").lower()

    if play == "yes":
        
        clear_output()

        game()

    else:

        break
```
