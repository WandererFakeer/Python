# Write a program that calculates the Body Mass Index (BMI) from a user's weight and height.

```python
def body_mass_index (height, weight):
    
    
    '''
        This function takes 2 float parameters, and calculates the Body Mass Index
        
    '''
    
    #1 Calculate the Body Mass Index by divinding the Weight with the square of the Height, and convert the result in a whole number
    bmi = round ( weight / (height ** 2) )
    
    #2. Returns the value
    return bmi
```

```python
def body_mass_index_explanation (height, weight):
    
    '''
        This function takes 2 parameters, and gives explanation according to the evaluated Body Mass Index
        
    '''
    
    #3. calls the "body_mass_index()" function, passing height and weight as 2 arguments, and saves the returned value to "index
    index = body_mass_index (height, weight)
    
    #4. Checks the conditions and provides explanation according to the evaluated index
    
    if (index < 18.5):
        print(f"Your BMI is {index}, you are underweight.")
        
    elif ( (index >= 18.5) and (index <= 25) ):
        print(f"Your BMI is {index}, you have a normal weight.")
        
    elif ( (index >= 25) and (index <= 30) ):
        print(f"Your BMI is {index}, you are slightly overweight.")
        
    elif ( (index >= 30) and (index <= 35) ):
        print(f"Your BMI is {index}, you are obese.")
        
    else:
        print(f"Your BMI is {index}, you are clinically obese.")
```

```python
#5. Ask the user about their Height in Meters
individual_height = float ( input ("Enter your height in Meters : ") )

#6. Ask the user about their Weight in KGs
individual_weight = float ( input ("Enter your weights in KGs : ") )

#7. Calls the function "body_mass_index_explanation()" passing the height and weight as 2 arguments
body_mass_index_explanation (height = individual_height, weight = individual_weight)
```
