```python
def game():

    while True:

        print('Welcome to Treasure Island!!!\nYour mission is to find the treasure.\n'
          'You\'re at a cross road. Where do you want to go?')

        direction = input('Type "Left" or "Right": ').lower()

        if (direction == "left"):

            print('You\'ve come to a lake.\n There is an island in the middle of the lake.\n  '
              'Type "Wait" to wait for a boat. Type "Swim" to swim across.')

            action = input().lower()

            if (action == "wait"):

                print('You arrive at the island unharmed.\n There is a house with 3 doors.\n  '
                  'One red, one yellow and one blue. Which colour do you choose?')

                door_color = input().lower()

                if (door_color == "red"):

                    print("It's a room full of fire.\n Game Over!!")

                elif (door_color == "yellow"):

                    print("You found the treasure!!\n You Win!!!")

                else:

                    print("You enter a room of beasts.\n Game Over!!")

            else:

                print("You get attacked by an angry trout.\n Game Over!!")

        elif (direction == "right"):

            print("You fell into a hole.\n Game Over!!")

        else:

            print('Your choice is invalid. Please type "Left" or "Right"')

        continuity = input('Do you want to continue the game? Type "Yes" or "No." ').lower()

        if (continuity == "no"):

            break
```
