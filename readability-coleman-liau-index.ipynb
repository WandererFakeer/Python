## Readability - Coleman-Liau index. "index = 0.0588 * L - 0.296 * S - 15.8"

```python
import re
```

```python
def total_letter(string):
    """

    :param string:
    :return: letter_count

    This function takes one parameter as string, and returns an integer

    """
    letter_count = 0

    for char in string:

        #checks if each of the character in "string" is an English alphabet
        if char.isalpha():
            letter_count += 1

    return letter_count
```

```python
def average(letter_length, sentence_length, word_length):

    """

    :param letter_length:
    :param sentence_length:
    :param word_length:
    :return: ave_letter, ave_sentence

    This function takes 3 integers, and returns a tuple of 2 floats

    """
    #the number of letters divided by the number of words, all multiplied by 100
    ave_letter = (letter_length / word_length) * 100

    # the number of sentences divided by the number of words, all multiplied by 100
    ave_sentence = (sentence_length / word_length) * 100

    return ave_letter, ave_sentence
```

```python
def calculation(average_letter, average_sentence):

    """

    :param average_letter:
    :param average_sentence:
    :return: None

    This function takes 2 float values, and returns None

    """
    #calculates the Coleman-Liau index
    index = round(0.0588 * average_letter - 0.296 * average_sentence - 15.8)

    if (index < 1):

        print("Before Grade 1")

    elif (index >= 16):

        print("Grade 16+")

    else:

        print(f"Grade {index}")
```

```python
def read():
    user_text = input("Enter a block of text to check the Grade : \n")

    # calls "total_letter()" to get the number of letters
    letter_l = total_letter(user_text)

    # finds the words with split() method
    word = user_text.split()

    # gets the number of words
    word_l = len(word)

    # finds the sentence, through regex method
    sentence = re.split('\!|\.|\?', user_text)

    # gets the number of sentence
    sentence_l = len(sentence) - 1

    # calls "average()" to get averages of letters and sentences
    letter_avg, sentence_avg = average(letter_l, sentence_l, word_l)

    # calls "calculation()" to get the Grade
    calculation(letter_avg, sentence_avg)
```

```python
while True:

    read()

    response = input("Do you want to continue checking? 'Yes' or 'No'? ")

    if response.lower() == "no":
        break
```
