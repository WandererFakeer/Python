```python
alphabets = [chr(num) for num in range(97, 123)]

alphabet_length = len(alphabets)
```

```python
def caeser(code_choice, user_message, shift_num):

    """

    :param code_choice:
    :param user_message:
    :param shift_num:
    :return: result

    This function takes 3 parameters, string user response of encode or decode,
    another string user message to be encoded or decoded, and integer shift number
    It returns the resultant encoded or decoded message

    """

    #empty string to be filled with encoded or decoded message
    result = ""

    #checks if "code_choice" is "decode"
    if code_choice == "decode":

        #converts the shift number into a negative number
        shift_num *= -1

    #traverses through the "user_message"
    for letter in user_message:

        #checks if each letter is in the "alphabets" list
        if letter in alphabets:

            #grabs the index position of each letter from the "alphabets" list
            position = alphabets.index(letter)

            #adds the index position of each letter to the shift number (for decoding, if will
             #be moved backward, instead of going forward, due to it being a negative number)
            position += shift_num

            #checks if the index position is greater or equal to "alphabet_length"
            if (position >= alphabet_length):

                #subtracts "alphabet_length" from "position" and reassigns the value to "position"
                position -= alphabet_length

            #grabs each letter from "alphabets" list and adds them to the empty string
            result += alphabets[position]

        else:

            result += letter

    return result
```

```python
def game():

    while True:

        code = input("Type 'Encode' to encrypt, type 'Decode' to decrypt:\n").lower()

        message = input("Type your message:\n").lower()

        shift = int(input("Type the shift number:\n"))

        #grabs the modulus of "shift" by "alphabet_length", and reassigns it to "shift"
        shift %= alphabet_length

        caeser_result = caeser(code, message, shift)

        print(f"Here's the {code}d result: {caeser_result}")

        continuity = input("Type 'Yes' if you want to go again. Otherwise type 'No': ").lower()

        if (continuity == "no"):

            print("Goodbye!")

            break
```
